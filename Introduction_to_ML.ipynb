````python
!pip install scikit-learn
````

````python
import pandas as pd

churn = pd.read_csv("telecom_churn_clean.csv")

churn.head()
````

````python
churn = pd.read_csv("telecom_churn_clean.csv", index_col = 0)

churn.head()
````

````python
X = churn[["total_day_charge", "total_eve_charge"]]

y = churn[["churn"]]

print(X.head())

print(y.head())

print("Data type of X and y: ", type(X), type(y))
````

````python
X = churn[["total_day_charge", "total_eve_charge"]].values # returns numpy array

y = churn["churn"].values # returns numpy array

print("Data type X and y: ", type(X), type(y))

print("Size of X and y: ", X.shape, y.shape)
````

````python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=15)

# training the model
knn.fit(X, y) 
````

````python
# testing on new data
X_new = np.array([[56.8, 17.5], [24.4, 24.1], [50.1, 10.9]]) 
````

````python
predictions = knn.predict(X_new)

# 1 = user churned
# 0 = subscription is intact
print(predictions)
print(type(predictions))
````

````python
for i in predictions:
  if (i == 1):
    print("churned")
  else:
    print("kept their subscription")
````
